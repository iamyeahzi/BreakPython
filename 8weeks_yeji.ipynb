# <B> 2. 데이터 시각화(EDA)를 위한 Matplotlib, Seaborn

### (1) Matplotlib, Seaborn 라이브러리 import

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

iris_df = sns.load_dataset("iris")
tips_df = sns.load_dataset("tips")

# 매직 라인
%matplotlib inline
```

### (2) <b>jointplot() :</b> 두가지의 그래프를 접목한 그래프 <br>(kind ="그래프 종류")를 통해 중앙 그래프를 결정

- <b>산점도 그래프(scatter) :</b> 두 변수간의 상관관계를 산점도로 표현<br>jointplot(x='변수명' , y='변수명', data=데이터명, <b> kind = 'scatter')

```python
jop1 = sns.jointplot(x='sepal_length', y = 'petal_length',  data=iris_df, kind = 'scatter') 
```

- <b>회귀 그래프(reg) :</b> 두 변수 간의 상관관계를 나타내고 회귀선을 그림<br>jointplot(x='변수명', y = '변수명', data=데이터명, <b> kind = 'reg')

```python
jop2 = sns.jointplot(x='sepal_length', y = 'petal_length', kind = 'reg', data=iris_df)
```

- <b>등고선 그래프(kde) :</b> 두 변수 간의 상관관계를 등고선(범위)로 나타냄, 빽빽할수록 데이터가 많음 <br>jointplot(x='변수명', y = '변수명', data=데이터명, <b> kind = 'kde')

```python
jop3 = sns.jointplot(x='sepal_length', y = 'petal_length', kind = 'kde', data=iris_df)
```

- <b>헥스 그래프(hex) :</b> 두 변수 간의 상관관계를 벌집 모양(밀도)으로 나타냄, 색이 진할수록 데이터가 많음<br>jointplot(x='변수명', y = '변수명', data=데이터명, <b> kind = 'hex')

```python
jop4 = sns.jointplot(x='sepal_length', y = 'petal_length', kind = 'hex', data=iris_df)
```

- <b>레지 그래프(resid) :</b> y=0 평균값을 중심으로 Residual(잔여값)를 그려줌<br>jointplot(x='변수명', y = '변수명', data=데이터명, <b> kind = 'resid')

```python
jop5 = sns.jointplot(x='sepal_length', y = 'petal_length', kind = 'resid', data=iris_df)
```

### (3) <b>countplot() :</b> 항목별 개수를 세어주는 그래프

- <b>세로로 그리기(x) :</b> <br>countplot(<b>x = '항목'</b>, data = 데이터명)

```python
sns.countplot(x = 'day', data = tips_df)
```

- <b>가로로 그리기(y) : </b> <br>countplot(<b>y = '항목'</b>, data = 데이터명)

```python
sns.countplot(y = 'day', data = tips_df)
```

### (4) <b>pairplot() :</b> 그리드(grid) 형태로 각 집합의 조합에 대해 히스토그램과 분포도를 그리는 그래프

```python
sns.pairplot(iris_df , hue="species")
```

### (5) <b>violinplot() :</b> 변수에 대한 데이터의 비교 분포도를 바이올린 모양으로 나타내는 그래프 <br>뚱뚱한 부분은 데이터의 분포를 나타내며, 양단의 뾰족한 부분은 데이터의 최소값과 최대값을 나타냄

- <b>한가지 변수에 대해 그리기(x) :</b> <br>violinplot<b>(x=데이터명["변수명"])</b>

```python
sns.violinplot(x=tips_df["total_bill"])
plt.show()
```

- <b>두가지 변수에 대해 그리기(x,y) :</b> <br>violinplot<b>(x=데이터명["변수명"], y=데이터명["변수명2"])</b>

```python
sns.violinplot(x="day", y="total_bill", data=tips_df)
```

### (6) <b>Implot() :</b> 변수 간의 상관관계를 나타내는 차트( 회귀선 포함 ) <br>이상치를 구별하는 것에 유용

```python
sns.lmplot(x="total_bill", y="tip", height=8, data=tips_df)
```
