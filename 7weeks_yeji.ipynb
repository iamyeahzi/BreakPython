# - <b>EDA( 데이터 시각화 )

> ## 1.   A사의 데이터 애널리스트의 입장에서, A사 음식점의 매출 데이터를 전달받아 EDA를 진행하려고 한다.

> ### - 데이터는 total_bill(계산금액), tip(팁 금액), sex(계산자 성별), smoker(테이블에 흡연자 포함여부), day(요일), time(시간대), size(테이블 인원 수)로 구성되어 있다.

----------------------------

### <b> 1. 데이터의 정보와 요약통계를 확인하라. (Head&Tail, 데이터 타입, Row&Columns 수, Std, Mean, NaN 수 등등 다다익선)

- <b> (1) head() : 상위데이터 출력( 5개 )

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
tips_dataframe = sns.load_dataset("tips")

tips_dataframe.head()
```

- <b>(2) tail() : 하위데이터 출력( 5개 )

```python
tips_dataframe.tail()
```

- <b>(3) mean() : ( 계산금액, 팁 금액, 인원수 )평균값 출력

```python
tips_dataframe[['total_bill','tip','size']].mean()
```

- <b>(4) groupby(['컬럼 이름']).mean() : 해당 컬럼별 평균값 출력

```python
# 성별(sex)별 평균값
tips_dataframe.groupby(['sex']).mean()
```

```python
# 인원수(size)별 평균값
tips_dataframe.groupby(['size']).mean()
```

```python
# 요일(day)별 평균값
tips_dataframe.groupby(['day']).mean()
```

```python
# 시간대(time)별 평균값
tips_dataframe.groupby(['time']).mean()
```

```python
# 흡연자여부(smoker)별 평균값
tips_dataframe.groupby(['smoker']).mean()
```

- <b>(6) std() : ( 계산금액, 팁 금액, 인원수 )표준 편차값 출력

```python
tips_dataframe[['total_bill','tip','size']].std()
```

- <b>(7) info() : 컬럼 별 데이터 타입 출력

```python
tips_dataframe.info()
```

- <b>(8) isnull() : 컬럼별 Nan 값 여부 출력

```python
tips_dataframe.isnull()
```

- <b>(9) isnull().sum() : 컬럼별 Nan 값( 결측값 ) 개수 출력

```python
tips_dataframe.isnull().sum()
```

- <b> (10) notnull().sum() : 컬럼별 실측값 개수 출력

```python
tips_dataframe.notnull().sum()
```

- <b>(11) notnull().sum(1) : 행 별 실측값 개수 출력

```python
tips_dataframe.notnull().sum(1)
```

### <b> 2.계산 금액과 팁 금액 사이에 어떠한 상관 관계가 있을 것 같다. 두 변수의 관계를 시각화하여 나타내고, 선형적 관계가 보이는지 회귀선까지 그어서 표시해라.

- <b>lmplot() : 컬럼 간의 선형관계를 확인하는 차트( 회귀선 포함 )

```python
# total_bill(계산 금액)과 tip(팁 금액) 사이의 선형관계 확인
sns.lmplot(x="total_bill", y="tip",data=tips_dataframe)
plt.show()
```

### <b> 3. 요일별로 팁의 발생 횟수가 어떻게 다른지를 시각화해서 나타내어 보아라. (Countplot 활용)

- <b> countplot() : 컬럼 별 데이터의 개수를 표시하는 차트

```python
sns.countplot(x='day',data=tips_dataframe)
plt.title("Daily tip")
plt.show()
```

### <b> 4. 요일별 계산금액의 통계적 분포를 한눈에 보고싶다. 시각화해서 나타내어 보아라.

- <b>stripplot() : 컬럼 별 데이터의 개수를 표시하는 차트

```python
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", data=tips_dataframe)
plt.title("daily total bills")
plt.show()
```

### <b> 5. 시간대, 요일, 테이블 인원수가 팁 금액과 어떠한 관계가 있는지 한눈에 알 수 있도록 시각화하라.

- <b> replot() : 컬럼 간 상관관계를 나타내는 차트 ( 회귀선 X )
    -  hue : 두가지 컬럼(변수) 간 상관관계를 나타낼 때, 또 하나의 변수를 색상 값을 통해 추가하여 나타냄
    -  col : 해당 컬럼(변수) 별로 차트를 생성

```python
sns.relplot(x="size", y="tip", hue="day", col="time", data=tips_dataframe)
plt.show()
```

### <b> 6. 추가적인 EDA를 진행할 수 있다면 진행하고, 얻어낸 Insight로 음식점의 매출액 혹은 팁 금액을 증대시키기 위한 방안을 제시하라.

- <b> replot() : 계산 금액과 팁과의 상관관계를 나타내고, 요일별로 분포를 표시한 뒤, 시간대별로 구분한 차트

```python
sns.relplot(x="total_bill", y="tip", hue="day",col="time", data=tips_dataframe)
plt.show()
```

### 결과 : 계산금액이 높을 수록 팁 금액 또한 높으며, 런치 시간에는 '목요일'이 디너 시간에는 '토요일'이 이러한 경향성을 뚜렷하게 가짐

----

- <b> barplot() : 요일별 계산금액의 평균을 나타내는 차트

```python
sns.barplot(x="day", y="total_bill", data=tips_dataframe)
plt.show()
```

### 결과 : 계산금액은 '일요일 > 토요일 > 목요일 > 금요일' 순으로 크다

---

- <b> barplot() : 시간대별 계산금액 / 팁 금액의 평균을 나타내는 차트

```python
sns.barplot(x="time", y="total_bill", hue="day", data=tips_dataframe)
plt.show()

sns.barplot(x="time", y="tip", hue="day", data=tips_dataframe)
plt.show()
```

### 결과 : 계산금액과 팁 금액 모두 런치시간대에는 '목요일 > 금요일' 순으로 높은 평균을 가지며, 디너시간대에는 '일요일 > 토요일' 순으로 높은 평균을 가진다.

> ### <b> 매출증대방안 

> ### 현재 A사 음식점은 평일에는 런치 위주의 매출과 주말에는 디너 위주의 매출이 생기고 있다. <br/> 팁 금액은 계산 금액이 높을수록 증가하는 경향을 보이기 떄문에, A사 음식점은 비교적 매출이 적은 평일 디너와 주말 런치의 매출을 높이는 방안을 찾아야 할 것이다. <br/> 평일 디너(18시-22시) 시간대에는 인근 회사원들을 대상으로 사원증 제시 시 할인을 적용해주는 방법을 이용하고, 주말 런치(12시-15시) 시간대에는 합리적인 가격의 런치 세트 메뉴를 개발하거나 특정 시간대(15-17시)에 해피아워 할인을 적용하여 매출을 증대 시키는 방법을 이용해 볼 수 있다. <br/> 매출이 부진한 평일 디너와 주말 런치를 활성화 시킨다면 A사의 음식점의 매출액은 증가할 것이다.
