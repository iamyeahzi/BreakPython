### (3) <b>dir_prev_bfnum(í•´ë‹¹ ê°ë…ì˜ ì´ì „ ì°¸ì—¬ ì˜í™” ê´€ê°ìˆ˜ì˜ í‰ê· )</b> ê³¼ <b>box_off_num(ê´€ê°ìˆ˜)</b> ì‚¬ì´ì˜ ê´€ê³„

- <b>ë‹¨ìˆœì„ í˜•íšŒê·€ë¶„ì„</b>(x = dir_prev_bfnum(í•´ë‹¹ ê°ë…ì˜ ì´ì „ ì°¸ì—¬ ì˜í™” ê´€ê°ìˆ˜ì˜ í‰ê· ), y= box_off_num(ê´€ê°ìˆ˜))

```python
X = movie_df[['dir_prev_bfnum']]
Y = movie_df[['box_off_num']] 
print(X, '\n\n')
print(Y)
```

- <b>train_test_split(ë…ë¦½ë³€ìˆ˜, ì¢…ì†ë³€ìˆ˜, test data ì‚¬ì´ì¦ˆ(%), ëœë¤ ì¶”ì¶œ ì‹œë“œê°’) :</b> datasetì„ training dataì™€ test dataë¡œ ë¶„í• 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5, random_state=10)

print('train data ê°œìˆ˜: ', len(X_train))
print('test data ê°œìˆ˜: ', len(X_test),"\n")
```

- <b>LinearRegression() : </b>ì„ í˜•íšŒê·€ë¶„ì„

```python
# sklearn ë¼ì´ë¸ŒëŸ¬ë¦¬ì—ì„œ ì„ í˜•íšŒê·€ë¶„ì„ ëª¨ë“ˆ ê°€ì ¸ì˜¤ê¸°
from sklearn.linear_model import LinearRegression

# ë‹¨ìˆœíšŒê·€ë¶„ì„ ëª¨í˜• ê°ì²´ ìƒì„±
lr = LinearRegression()

# train dataë¥¼ ê°€ì§€ê³  íšŒê·€ëª¨ë¸ í•™ìŠµ
lr.fit(X_train, Y_train)

# í•™ìŠµì„ ë§ˆì¹œ ëª¨í˜•ì— test dataë¥¼ ì ìš©í•˜ì—¬ ê²°ì •ê³„ìˆ˜(R^2) ê³„ì‚°
r_square = lr.score(X_test, Y_test)

# íšŒê·€ì‹ê³¼ ê²°ì •ê³„ìˆ˜(R^2) ì‚°ì¶œ
print('íšŒê·€ì‹ :', float(lr.coef_),'X +',lr.intercept_)
print('ê²°ì •ê³„ìˆ˜(R^2) :', r_square)
```

> #### ğŸ”<b>ê²°ì •ê³„ìˆ˜ê°€ 0.1ë¡œ ê±°ì˜ ë¬´ì˜ë¯¸í•œ ì„ í˜•ê´€ê³„ë¼ê³  ë³¼ ìˆ˜ ìˆë‹¤.
